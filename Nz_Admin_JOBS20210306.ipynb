```python
import pandas as pd
import re
```

```python
def load_data(file_name):
    return pd.read_excel(file_name,engine="openpyxl",header = None)
df = load_data("NZ_Admin_JOBS.xlsx")
df.columns = ["job_name","link","3","4","5","6"]
df = df.fillna(value="Private Advertiser").rename(columns = {"3":"Company"})
df
```

```python
df['ad_type'] = df['link'].str.extract('\=(\w+)\#')
df['ad_type']
```

```python
df['4'] = df['4'].str.replace(" ","")
df['4']
```

```python
df['location'] = df['4'].str.extract('\:(\w+)area')
df['location']
```

```python
df['area'] = df['4'].str.extract('(\w+)$')
df['area']
```

```python
df['5'] =df['5'].str.replace(" ","")
df['list_time'] = df['5'].str.extract('(\w+),')
df['list_time']
```

```python
df['6'] =df['6'].str.replace(" ","")
df['classification'] = df['6'].str.extract(':(\w+)$')
df['classification']
```

```python
del df['4']
```

```python
del df['5']
del df['6']
```
